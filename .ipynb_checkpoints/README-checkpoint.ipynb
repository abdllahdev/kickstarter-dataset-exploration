# Introduction
***
In this project, we, Team Failed Successfully, aim to study and investigate a dataset containing data from the famous crowdfunding webpage "Kickstarter". Kickstarter is an Internet based funding platform in which ambitious, innovative, and imaginative ideas can be brought to life by connecting project creators with backers, that can directly support them. This platform, as many others, comes from the idea of "crowdfunding", the process of funding a project by raising small amounts of money from a large number of people. This practice has become really popular in the last 15 years, and since the launch of Kickstarter in 2009, the total crowdfunding revenue has hugely increased thorugh the years. Crowdfunding revenue tripled from \$530 million in 2009 to $1.5 billion in 2011, and in 2015 it raised to \$34 billion.


## Why explore the Kickstarter database?

Over the last 15 years, the rise of Kickstarter has allowed a democratisation of ideas by allowing anyone to realize their dreams of running a business or bringing a product to market. Kickstarter allows the enthusiasm of the general public to dictate whether an idea should be funded rather than a select group of investors who don’t necessarily know what people want. However, with this exciting opportunity comes a new arena in which to sell one’s idea. Instead of a business pitch, businesses must negotiate the art of setting their target goal, compelling the public to back their project and advertising to the right people to successfully achieve thier funding goal. We will, therefore, be analysing this dataset to get a better understanding of how to make a good Kickstarter campaign and therefore, we will be able  to make recommendations for future companies to improve their chances of success.

## Cleaning the dataset
Sicne we are interested in finding the factors that make the projects successful or failed, we are going to exclude these columns from the dataset (currency, goal, pledged and usd_pledged) and instead we will work on the usd_pledged_real and usd_goal_real. Because the values of these two columns are more reliable than the usd_pledged as stated in the description of the dataset on kaggle. Also, we dropped the goal and pledged because these columns have information about the money in the local currencies and we want to use only one currency so we can evalute the projects.

```python
ks = ks.drop(columns=['currency', 'goal', 'pledged', 'usd pledged'])
```

Also, we are not interested we are just interested in only two states which are successful and failed we are going to exclude all the other states

```python
only_successful_fail_projects = (ks['state'] == 'successful') | (ks['state'] == 'failed')
ks = ks[only_successful_fail_projects]
```

Moreover, our project will focus on the projects posted from 2009 to 2017 because the dataset doesn't have enough data about projects posted in 2018.

```python
ks['launch_year'] = pd.DatetimeIndex(ks['launched']).year
projects_2009_2017 = (ks['launch_year'] >= 2009) & (ks['launch_year'] <= 2017)
ks = ks[projects_2009_2017]
```

```python
# add a new column to represent the state as 0 or 1
ks["state_0_1"] = ks["state"].apply(lambda x: 1 if x == "successful" else 0)
```

```python
# remove the projects which have goal less that 10$ and more that 10,000,000$
ks = ks[(ks['usd_goal_real'] >= 10) & (ks['usd_goal_real'] <= 10000000)]
```

### Successful vs Failed projects

```python
successful_projects_n = np.count_nonzero(ks['state'] == 'successful')

failed_projects_n = np.count_nonzero(ks['state'] == 'failed')

total = successful_projects_n + failed_projects_n
```

Now, with this data, we can know the numerical relationship between the successful and failed projects. More specifically, we can know, for example, what the percentage of failed projects in our dataset is:

```python
percentage_successful = (failed_projects_n/total)*100
print("The percentage of successful projects: "+str(round(percentage_successful, 2))+'%')
percentage_failed = (failed_projects_n/total)*100
print("The percentage of failed projects: "+str(round(percentage_failed, 2))+'%')
```

Now, we can also observe how this number fluctuates as time passes. For this, we can extract the number of failed and successful projects for each year and plot them as shown below:

```python
failed_projects_per_year = []
years = np.array([year for year in ks.groupby("launch_year").groups.keys()])

for n in years:
    loop_failed_projects_per_year = np.count_nonzero((ks['launch_year']==n) & (ks['state']==('failed')))
    failed_projects_per_year.append(loop_failed_projects_per_year)

successful_projects_per_year = []

for n in years:
    loop_successful_projects_per_year = np.count_nonzero((ks['launch_year']==n) & (ks['state']==('successful')))
    successful_projects_per_year.append(loop_successful_projects_per_year)
```

```python
fig = plt.figure(figsize=(18,7))
plt.bar(years-0.15, failed_projects_per_year, width=0.3, color=(0.9, 0, 0, 0.8), label='Failed')
plt.bar(years+0.15, successful_projects_per_year, width=0.3, color=(0, 0.6, 0, 0.8), label='Successful')
plt.title('Number of failed and successful projects per year')
plt.xlabel('Year')
plt.ylabel('Number of projects')
plt.legend()
```

As seen in the graph, since the launch of Kickstarter in 2009, the number of successful and failed projects was approximately even until the year 2014. Then, the number of failed projects increased significantly while the number of successful projects has stayed approximately constant over the last 6 years of data. This could mean that, even though the total number of projects increases over time. However, this does not mean that the total amount of money backed has been constant too, since we don't know what the goals of the successful projects were.

```python
np.count_nonzero(just_success['usd_goal_real'] <= 10)
```

In fact, only 200 of the successful projects had a goal of $10 of under. 

Therefore it is not as simple, as just setting the goal really low to be successful. There are many reasons why this might be the case such as: 
- People many not take a business or idea seriously if the goal is set too low as they may think it is a joke or not feasible.
- By guaranteeing success, people may be more sceptical of the project as their money will definately be taken e.g. a project with a goal of \$10 but realistically needs \$10,000 will still be taking peoples' money even if only  \$1000 is raised (and can therefore not happen) compared to setting a goal of \$10,000 where the money will only be taken if the project definately can take place.
- People may be attracted to the idea of a large goal, both as a motivational tool to rally people around a cause (they may be more likley to get friends and family on board) and also may be more likely to attract media attention.
- There is less control over the total amount raised by the project, which has more of an effect on the outcome of the project e.g. a project expecting \$5000 but only getting \$100 is still obligated to deliver. It therefore, may not be a risk businesses are willing to take.

# Finding the correlation between the success and the failiure of the projects and the different attributes
In this part of the project we are going to look at the correlation between different variables and the success of the project. This helps us determine the key factors to consider when creating a project and how influential they are on the success of the project.

## The correlation between the main-category and the sub-category and the success of the project
First let's find the number of failed and successful projects in each main category

### Number of successful vs failed projects in each main category

```python
# percentage of successful projects in each main category
successful_projects_main_categories = ks['state'] == 'successful'
successful_projects_main_categories = ks[successful_projects_main_categories]
successful_projects_main_categories = (successful_projects_main_categories.groupby('main_category').size()/main_categories_projects) * 100

# percentage of failed projects in each main category
failed_projects_main_categories = ks['state'] == 'failed'
failed_projects_main_categories = ks[failed_projects_main_categories]
failed_projects_main_categories = (failed_projects_main_categories.groupby('main_category').size()/main_categories_projects) * 100

successful_projects_main_categories
```

Plotting the number of failed projects vs the number of successfull project to see if the categories affect the failure or the success of the project:

```python
# location of bars on x-axis
x = np.arange(main_categories_count)

# bars width
width = 0.35

fig, ax = plt.subplots(figsize=(18, 6))
successful_bar = ax.bar(x-(width/2), successful_projects_main_categories, width, label='Successful', color='green')
failed_bar = ax.bar(x+(width/2), failed_projects_main_categories, width, label='Failed', color='red')
ax.set_title('Percentage of failed vs successful projects in each main category')
ax.set_xticks(x)
ax.set_xticklabels(successful_projects_main_categories.groupby('main_category').groups)
ax.legend()

plt.show()
```

From the above graph, we can see that the success rate in some categories like Theater, Dance, and Comics is higher than the failure rate. Also, some of the categories' failure rate is more than the double in categories like Technology, Journalism, Food, and Crafts. This might incidate that there is a strong relation between the main category and the success of the project.

Now, let's calculate the correlation between the main category and the success of the project

```python
# calculate the correlation between backing period and success of the project by using pandas corr() method.
# we are using spearman method because we are finding the correlation between two categorical variables
ks['main_category'].corr(ks['state_0_1'], method='spearman')
```

Unlike our prediction from the graph, the correlation coeffiectant indicate taht there is no relation between the main category and the success of the project

# Create a model to predict if a project is going to fail or success
In this section of the project, we are going to build a logistic regression to predict if a project is likely to be successful or not. We are going to create our model using the following variables: launch month, main-category, sub-category, country, goal group, and backing period.

We are building a logistic regression model because most of the variables are categorical.

## Cleaning the dataset for the model

```python
# sigmoid function to normalize the data of the backers and the backing period to make its values between 0 and 1
def sigmoid1(x):
    return 1/(1 + np.exp(-x))
```

```python
# create a new table for the model
ks_copy = pd.DataFrame({'state': ks['state_0_1'],
                        'main_category': ks['main_category'],
                        'sub_category': ks['category'],
                        'country': ks['country'],
                        'goal_group': ks['goal_group'],
                        'launch_month': ks['launch_month'],
                        'backing_period': ks['backing_period'],
                        'backers': ks['backers']})

# normalize the backer and backing period columns
ks_copy['backers'] = sigmoid1(ks_copy['backers']/ks_copy['backers'].mean())
ks_copy['backing_period'] = sigmoid1(ks_copy['backing_period']/ks_copy['backing_period'].mean())

# convert categorical variable into indicator variables
ks_copy = pd.get_dummies(ks_copy,
                         columns=['main_category', 'sub_category', 'country', 'goal_group', 'launch_month'],
                         prefix=['main_category', 'sub_category', 'country', 'goal_group', 'launch_month'])

ks_copy.head()
```

## Creating the model

```python
# method to calculate sigmoid function
def sigmoid2(x, weights):
    z = np.dot(x, weights[1:]) + weights[0]
    return 1/(1 + np.exp(-z))

# cost funtion
def cost(y, prediction):
    return -y.dot(np.log(prediction)) - ((1 - y).dot(np.log(1-prediction)))
```

```python
# the gradient descent function
# we use this function to find the weights which are the factors of the variables of the decision boundry function. The decision boundry is the regression function
def gradient_descent(alpha, weights, epochs, training_set):
    y = training_set.iloc[:len(training_set), 0].values
    x = training_set.iloc[:len(training_set), 1:].values

    for i in range(epochs):
        prediction = sigmoid2(x, weights)
        error = prediction - y
        gradient = x.T.dot(error)

        weights[0] = weights[0] - alpha * error.sum()
        weights[1:] = weights[1:] - alpha * gradient

    return {'cost': cost(y, prediction), 'weights': weights}
```

```python
m, n = ks_copy.shape
gd = gradient_descent(0.00001, np.zeros(n), 100, ks_copy[:150000])
```

```python
def predict(x, weights):
    return 1 if sigmoid2(x, weights) >= 0.5 else 0
```

```python
# test the classifier and calculate the accuracy
predictions = np.array([])

test_dataset = ks_copy.iloc[150000: ,1:].values

for i in test_dataset:
    predictions = np.append(predictions, predict(i, gd['weights']))
```

```python
accuracy = predictions == ks_copy.iloc[150000: ,0].values
np.count_nonzero(accuracy)/len(predictions)
```

- Although some may be put off by the fact that there has been a slight decline from peak popularity from around 2014- 2015 (https://trends.google.com/trends/explore?date=all&q=kickstarter) the total amount of successful projects funded has remained roughly stable from 2012 onwards.


- Advice for product categories: Technology has the highest average goal per project but also only receives 1/3 of the goal on average, therefore it maybe beneficial for tech companies to set more realistic goals if they want to get funded.
